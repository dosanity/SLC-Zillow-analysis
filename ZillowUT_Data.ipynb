```python
df = pd.DataFrame()

zpid = []
price = []
address = []
city = []
state = []
zipcode = []
home_type = []
beds = []
baths = []
area = []
latitude = []
longitude = []
broker = []
site_url = []

for i in range(len(json_data)):
    
    listResults = json_data[i]['cat1']['searchResults']['listResults']
    
    for j in range(len(listResults)):
        
        try:
            zpid.append(listResults[j]['id'])
        except:
            zpid.append(np.nan)
        
        try:
            price.append(listResults[j]['unformattedPrice'])
        except:
            price.append(np.nan)
            
        try:
            address.append(listResults[j]['addressStreet'])
        except:
            address.append(np.nan)
        
        try:
            city.append(listResults[j]['addressCity'])
        except:
            city.append(np.nan)
            
        try:
            state.append(listResults[j]['addressState'])
        except:
            state.append(np.nan)
            
        try:
            zipcode.append(listResults[j]['addressZipcode'])
        except:
            zipcode.append(np.nan)
            
        try:
            home_type.append(listResults[j]['hdpData']['homeInfo']['homeType'])
        except:
            home_type.append(np.nan)
            
        try:
            beds.append(listResults[j]['beds'])
        except:
            beds.append(0)

        try:
            baths.append(listResults[j]['baths'])
        except:
            baths.append(0)

        try:
            area.append(listResults[j]['area'])
        except:
            area.append(np.nan)
            
        try:
            latitude.append(listResults[j]['latLong']['latitude'])
        except:
            latitude.append(np.nan)
            
        try:
            longitude.append(listResults[j]['latLong']['longitude'])
        except:
            longitude.append(np.nan)
            
        try:
            site_url.append(listResults[j]['detailUrl'])
        except:
            site_url.append(np.nan)
            
        try:
            broker.append(listResults[j]['brokerName'])
        except:
            broker.append(np.nan)
            
df['id'] = zpid
df['price'] = price
df['address'] = address
df['city'] = city
df['state'] = state
df['zipcode'] = zipcode
df['type'] = home_type
df['beds'] = beds
df['baths'] = baths
df['sqft'] = area
df['latitude'] = latitude
df['longitude'] = longitude
df['url'] = site_url
df['broker'] = broker

df = df.drop_duplicates()
```
